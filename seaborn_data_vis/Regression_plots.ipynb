```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,hue='time', markers=['o', 'v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time', row='sex',
          aspect=0.8, size=8)
```
